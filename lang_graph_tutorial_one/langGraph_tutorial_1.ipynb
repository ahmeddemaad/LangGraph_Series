<div class="langgraph-tutorial">
  <h2>LangGraph Overview</h2>
  <p>LangGraph is a robust and versatile tool designed for building <strong>stateful</strong>, <strong>multi-actor</strong> applications with Large Language Models (LLMs). Built as an extension of the LangChain library, LangGraph enhances your ability to create complex, dynamic, and collaborative workflows.</p>

  <h3>Key Features:</h3>
  <ul>
    <li><strong>Stateful Design</strong><br>LangGraph maintains and understands the state of each node within a cycle. This ensures that the flow of information and decisions is context-aware and dynamically adjusted as the process evolves.</li>
    <li><strong>Multi-Actor Coordination</strong><br>LangGraph excels in coordinating multiple independent actors. These actors could include:
      <ul>
        <li>Different language models.</li>
        <li>Chains of processes.</li>
        <li>Computational components.</li>
      </ul>
      Each actor performs specific roles, enabling the development of sophisticated, modular, and scalable applications.
    </li>
  </ul>

  <h3>Use Cases:</h3>
  <p>LangGraph is ideal for building complex systems that require seamless integration and collaboration among multiple LLMs and processing units, such as:</p>
  <ul>
    <li>Advanced conversational agents.</li>
    <li>Automated decision-making systems.</li>
    <li>Multi-component language processing workflows.</li>
  </ul>

  <p>With LangGraph, you can harness the full potential of LLMs to create intelligent, state-aware, and cooperative applications with ease.</p>
</div>

<div class="langgraph-tutorial">
  <h2>Key Concepts</h2>

  <ul>
    <li><strong>Stateful Graph</strong><br>LangGraph revolves around the concept of a stateful graph, where each node in the graph represents a step in your computation, and the graph maintains a state that is passed around and updated as the computation progresses.</li>
    <li><strong>Nodes</strong><br>Nodes are the building blocks of your LangGraph. Each node represents a function or a computation step. You define nodes to perform specific tasks, such as processing input, making decisions, or interacting with external APIs.</li>
    <li><strong>Edges</strong><br>Edges connect the nodes in your graph, defining the flow of computation. LangGraph supports conditional edges, allowing you to dynamically determine the next node to execute based on the current state of the graph.</li>
  </ul>
</div>

<div class="coding-tutorial">
  <h3>Enhancing the <strong>messages</strong> Attribute with <strong>operator.add</strong></h3>
  <p>The <code><strong>messages</strong></code> attribute within the <code><strong>state</strong></code> can be enhanced by incorporating <code><strong>operator.add</strong></code> in the annotation. This enables the seamless addition of new messages to the existing list using the <code><strong>+</strong></code> operator, allowing the <strong>messages</strong> list to grow dynamically as new messages are added.</p>

  <p>This approach is particularly beneficial for maintaining a <strong>message history</strong>. Each new message is automatically appended to the list, creating a <strong>complete record</strong> of all past interactions. This ensures the application retains a <strong>chronological log</strong>, which can be invaluable for <strong>tracking</strong>, <strong>debugging</strong>, or enabling <strong>context-aware processing</strong>.</p>

  <blockquote>
    <strong>Note:</strong> This is <strong>not</strong> shared memory across different queries. Instead, it’s a memory shared within nodes, allowing state persistence during interactions but <strong>not</strong> across different query executions.
  </blockquote>
</div>

```python
from typing import TypedDict, Annotated, List
# through operators we can choose (add)
import operator

class State(TypedDict):
    messages: Annotated[List[str], operator.add]
```

  <h2>Step 2: Define Nodes</h2>

  <p>In LangGraph, we need to define a few key nodes for our graph. Each node can be either a function or a runnable. Here are the two primary nodes required:</p>

  <ul>
    <li><strong>Agent Node</strong><br>Responsible for determining if any actions should be taken, and, if so, deciding which actions are appropriate.</li>

  <p>These nodes work together to enable decision-making and action execution within the graph.</p>

```python
from langgraph.graph import StateGraph ,END,START

# Thats the function which would be mapped to a node later
def add_message(state):
    # we are going to print the state once we pass through this node
    print(state)
    new_msg= "hello"
    # The return here means the output is going to be passed to next node (Also it's going to be appended in our temp. memory)
    return{"messages" : [new_msg]}

# Thats another the function which would be mapped to a node later also :)
def add_another_message(state):
    print(state)
    new_msg= "world"
    return{"messages" : [new_msg]}

# bulding graph
builder = StateGraph(State)

# mapping functions to nodes
builder.add_node("my_node_1",add_message)
builder.add_node("my_node_2",add_another_message)
```

<div class="langgraph-tutorial">
  <h2>Step 3: Connect the Edges</h2>

  <p>Edges are the connectors between nodes, defining the flow and logic of the graph. They enable nodes to interact and create a cohesive workflow. In LangGraph, there are two types of edges:</p>

  <ul>
    <li><strong>Normal Edge</strong><br>
      <ul class="nested-list">
        <li><strong>Purpose:</strong> Ensures that after tools are invoked, the flow always returns to the agent to decide the next steps.</li>
        <li><strong>Usage:</strong> Creates a consistent loop for continuous decision-making and action execution.</li>
      </ul>
    </li>
    <li><strong>Conditional Edge</strong> (Explored in more detail later)<br>
      <ul class="nested-list">
        <li><strong>Purpose:</strong> Introduces branching logic based on the agent's response.</li>
        <li><strong>Behavior:</strong>
          <ul class="nested-list">
            <li>If the agent determines that an action should be taken, the flow proceeds to the tool invocation function.</li>
            <li>If the agent decides the task is complete, the flow terminates.</li>
          </ul>
        </li>
      </ul>
    </li>
  </ul>

  <h3>Why Edges Matter</h3>
  <p>Edges establish the structure of the graph by:</p>
  <ul>
    <li>Defining the sequence of operations.</li>
    <li>Introducing conditional logic to create dynamic, state-aware workflows.</li>
  </ul>

  <p>By carefully connecting edges, you shape the behavior and decision-making process of your LangGraph application.</p>
</div>

```python
# Setting the entry point of the graph (i.e., the starting point of the graph)
builder.add_edge(START,"my_node_1")

# Connecting node 1 and node 2 with a direct edge (you can think of it as a straight line between them)
builder.add_edge("my_node_1","my_node_2")

# Adding an edge from "my_node_2" to the designated END node (where our graph ends)
builder.add_edge("my_node_2",END)

# compiling the graph to start using it
graph = builder.compile()
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage

# Step 1 : Define Model we are going to use from openAi
llm = ChatOpenAI(model="gpt-4o", api_key=os.environ.get("OpenAi_Key"))

# Step 2.1 : Define State
class State(TypedDict):
    messages:Annotated[list[str],operator.add] # remember we use add operator as a temp memory for now .

# Step 2.2 : build state    
builder = StateGraph(State)

# Step 3.1 : build functions (which are going to be nodes)
def call_model(state):
    ''' 
    This function is going to take the message of the user and pass it to the model to answer user question 
    and receive the resoponse from the model and return it (returning here means : passing the output to the next node automatically )
    '''
    print("state of Model call",state)
    # get all messages in messages lists inside state
    messages = state["messages"]
    # we pass it to the model to get (inorder for the model to preserve history and to have context understanding)
    response = llm.invoke(messages)
    return {"messages":[response]}

def convert_human(state):
    """
    Just taking the data adding it in other object called humanMessage just to show how easy you can control the inputs/output values
    """
    print("state of Conv Human",state)
    last_message = state["messages"][-1]
    content = last_message.content
    response = HumanMessage(content=content)
    return{"messages":[response]}

# Step 3.2 : Add Assign Nodes to a specific name
builder.add_node("call_model",call_model)
builder.add_node("convert_human",convert_human)

# Step 4 : Set Graph
builder.set_entry_point("call_model") # similar to builder.add_edge(START,"call_model")
builder.add_edge("call_model","convert_human")
builder.add_edge("convert_human",END)

graph = builder.compile()
```

<div class="langgraph-tutorial">
  <h2>Why State is Essential in LangGraph for Tool Usage</h2>
  <p>
    It is <strong>effectively impossible</strong> to use tools in LangGraph without managing state because the state is integral to:
  </p>
  <ul>
    <li><strong>Storing the results of tool calls</strong>: The output of each tool call needs to be captured and passed along to the next step.</li>
    <li><strong>Providing context to the LLM</strong>: LLMs require context to generate relevant and meaningful responses based on prior tool outputs.</li>
    <li><strong>Managing the flow of execution across the graph</strong>: State ensures that the sequence of tool invocations and LLM calls follows a logical order.</li>
  </ul>
  <p>
    Without state, the tools cannot communicate properly, and the graph's execution flow would break down.
  </p>
</div>

```python
from langgraph.prebuilt import ToolNode
from langgraph.graph import Graph
from langchain_core.messages import HumanMessage, AIMessage
from langchain_openai import ChatOpenAI
from langgraph.prebuilt import ToolNode
```

```python
tools = [OpenWeatherMapQueryRun(),predict_age]
```

```python
class our_state(TypedDict):
    messages: Annotated[list[str], operator.add]
```

```python
LLM = ChatOpenAI(temperature=0,model='gpt-4o')
LLM = LLM.bind_tools(tools)
tool_node = ToolNode(tools)
```

```python
def trigger_LLM(state):
    response = LLM.invoke(state['messages'])
    return {"messages": [response]}
```

```python
def should_continue(state):
    messages = state['messages']
    last_message = messages[-1]
    if last_message.tool_calls:
        return "tools"
    
    return END
```

```python
workflow = StateGraph(our_state)
```

```python
workflow.add_node("agent", trigger_LLM)
workflow.add_node("tools", tool_node)

workflow.add_conditional_edges(

    "agent", 
    should_continue
)
workflow.set_entry_point("agent")
workflow.add_edge("tools", 'agent')
```

    <ul style="list-style-type: disc; padding-left: 20px; color: black;">
        <li><strong>Map</strong>: The input data is divided into smaller tasks, which are processed in parallel </li>

 We'll use a state graph to process each topic individually (Map) and then aggregate the results (Reduce). The core idea is to generate information about multiple topics in parallel and then combine the results. </h3>

```python
def get_Topics(state):
    print(state['topics'])
    return {"topics":state['topics'] }
    
def generate_info(state):
    topic=state['topics']
    print(topic)
    response=LLM.invoke(f"Generate me any single random info about this Topic :{topic}")
    return {"results": [response]}
    
```

<div style="background-color: #2C3E50; color: #ECF0F1; padding: 20px; border-radius: 8px; box-shadow: 0 4px 8px rgba(0, 0, 0, 0.2); font-family: 'Arial', sans-serif; max-width: 600px; margin: 20px auto; text-align: center;">
  <h2 style="color: #1ABC9C; margin-bottom: 10px;">Send() Function</h2>
  <p>The <span style="font-weight: bold; color: #F39C12;">Send()</span> function dynamically generates new tasks at runtime, routing execution to specific nodes. Each <span style="font-weight: bold; color: #F39C12;">Send</span> action creates a new task, which is executed in <span style="font-weight: bold; color: #F39C12;">parallel</span> (or sequentially based on the logic).</p>
  <p><span style="font-weight: bold; color: #F39C12;">Each task</span> generated by <span style="font-weight: bold; color: #F39C12;">Send()</span> is typically isolated in its own execution context. This ensures that tasks can run independently without interfering with each other, providing greater flexibility and control in managing workflows.</p>
</div>


```python
def continue_to_Topics(state):
     return [Send("generate_info", {"topics": s}) for s in state["topics"]]
```

```python
workflow2 = StateGraph(my_state)

workflow2.add_node("get_Topics", get_Topics)
workflow2.add_node("generate_info", generate_info)

workflow2.add_conditional_edges(

    "get_Topics",
    continue_to_Topics,['generate_info'])
workflow2.set_entry_point("get_Topics")
workflow2.set_finish_point("generate_info")
```

```python
app = workflow2.compile()
```

    <ul style="list-style-type: disc; padding-left: 20px; color: #004d40;">
        <li><strong>Start:</strong> Execution begins with an empty <code>msgs</code> list.</li>

        <li><strong>End:</strong> The execution finishes, and the final state of <code>msgs</code> is: <code>[1, 2, 3, 6]</code>.</li>

    <h3 style="color: #00796b;">Key Insights:</h3>
    <ul style="list-style-type: disc; padding-left: 20px; color: #004d40;">
        <li><strong>Parallel Execution:</strong> Nodes "2" and "3" are executed in parallel, meaning they perform their tasks simultaneously.</li>
        <li><strong>State Update Order:</strong> Despite executing in parallel, <code>Node "2"</code>'s result (the value <code>[2]</code>) is appended to <code>msgs</code> before <code>Node "3"</code>'s result (the value <code>[3]</code>) because of the order in which the nodes were added to the graph.</li>

    </ul>
    <p style="color: #004d40;"><strong>Note:</strong> Even though nodes "2" and "3" were executed in parallel, the order in which their results are appended to the state depends on their creation order in the graph, not their execution order.</p>
